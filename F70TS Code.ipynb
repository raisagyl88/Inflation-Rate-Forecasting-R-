# Question 5

**Train and Test Data**

```python
from sklearn.model_selection import train_test_split

# Separate the features and the target
X = data.drop(columns=['Class']).values# Drop the 'Class' column to get the feature matrix
y = data['Class'].values# Select the 'Class' column as the target vector

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

```

```python
import numpy as np
from collections import Counter


class KNN:
    def __init__(self, k=3):
        self.k = k

    def fit(self, X, y):
        self.X_train = X
        self.y_train = y

    def predict(self, X):
        predictions = [self._predict(x) for x in X]
        return predictions

    def _predict(self, x):
        # compute the distances
        distances = np.sqrt(np.sum((self.X_train-x) ** 2, axis=1))

        # get the closest k neighbors
        k_indices = np.argsort(distances)[:self.k]
        k_nearest_labels = [self.y_train[i] for i in k_indices]

        # majority vote
        most_common = Counter(k_nearest_labels).most_common()
        return most_common[0][0]


```

```python
#Implementing kNN with k=3
knn=KNN(k=3)
knn.fit(X_train, y_train)

# Generate predictions
predictions = knn.predict(X_test)

from sklearn.metrics import classification_report, accuracy_score, f1_score

# Classification report
print("Classification Report:\n", classification_report(y_test, predictions))
```

# Question 6

**Tuning of hyperparameter k**

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, accuracy_score, f1_score, confusion_matrix
import matplotlib.pyplot as plt

# Initialize lists to store accuracy and F1 scores for each k
test_accuracies = []
test_f1_scores = []

# Loop through each value of k from 1 to 10
k_values = range(1, 11)

for k in k_values:
    print(f"\nEvaluating kNN model with k={k}")
    
    # Create and train the kNN model
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    
    # Make predictions on the test set
    y_pred = knn.predict(X_test)
    
    # Calculate Accuracy score and append to list
    accuracy = accuracy_score(y_test, y_pred)
    test_accuracies.append(accuracy)
    print(f"Accuracy on test data: {accuracy}")

    # Calculate F1 score and append to list
    f1 = f1_score(y_test, y_pred, average='weighted')
    test_f1_scores.append(f1)
    print(f"F1 Score on test data: {f1}")

    # Print Classification Report and Confusion Matrix
    print("Classification Report:\n", classification_report(y_test, y_pred))
    conf_matrix = confusion_matrix(y_test, y_pred)
    print(f"Confusion matrix:\n {conf_matrix}")

# Plotting Accuracy and F1 Score vs. k
plt.figure(figsize=(12, 5))

# Plot Test Accuracy
plt.subplot(1, 2, 1)
plt.plot(k_values, test_accuracies, marker='o', color='blue')
plt.xlabel("k")
plt.ylabel("Test Accuracy")
plt.title("Test Accuracy vs. k in kNN")

# Plot Test F1 Score
plt.subplot(1, 2, 2)
plt.plot(k_values, test_f1_scores, marker='o', color='orange')
plt.xlabel("k")
plt.ylabel("Test F1 Score")
plt.title("Test F1 Score vs. k in kNN")

plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier

# Define the parameter grid correctly
parameters = {'n_neighbors': range(1, 11)}

# Set up GridSearchCV with KNeighborsClassifier
knn_grid = GridSearchCV(KNeighborsClassifier(), parameters, n_jobs=4, cv=5)
knn_grid.fit(X=X_train, y=y_train)

# Get the best model and results
knn_model = knn_grid.best_estimator_
print("Best kNN accuracy:", knn_grid.best_score_, "-> Best k value:", knn_grid.best_params_)
```

# Question 7

**decision tree algorithm**

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, accuracy_score, f1_score, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Define depth values for the Decision Tree model
d_values = range(1, 11)

# Lists to store accuracy and F1 scores for plotting
test_accuracies = []
test_f1_scores = []

# Loop through each value of depth (d) from 1 to 10
for depth in d_values:
    print(f"\nEvaluating Decision Tree model with depth={depth}")
    dt = DecisionTreeClassifier(max_depth=depth, random_state=42)

    # Fit the model on the training data
    dt.fit(X_train, y_train)
    dt_preds = dt.predict(X_test)
    
    # Print the classification report
    print("Classification Report:\n", classification_report(y_test, dt_preds))
    
    # Calculate and print the accuracy on the test data
    dt_test_acc = accuracy_score(y_test, dt_preds)
    print(f"Accuracy on test data: {dt_test_acc}")
    test_accuracies.append(dt_test_acc)
    
    # Calculate and print the F1 Score on the test data
    dt_test_f1 = f1_score(y_test, dt_preds, average='weighted')
    print(f"F1 Score on test data: {dt_test_f1}")
    test_f1_scores.append(dt_test_f1)
    
    # Compute and print the confusion matrix
    dt_conf_matrix = confusion_matrix(y_test, dt_preds)
    print(f"Confusion Matrix:\n{dt_conf_matrix}")

    # Plot the confusion matrix
    plt.figure(figsize=(6, 4))
    sns.heatmap(dt_conf_matrix, annot=True, fmt="d", cmap="Blues", xticklabels=["Class 0", "Class 1"], yticklabels=["Class 0", "Class 1"])
    plt.xlabel("Predicted Labels")
    plt.ylabel("True Labels")
    plt.title(f"Confusion Matrix for Decision Tree (Depth={depth})")
    plt.show()

# Plot Test Accuracy and F1 Score for different depths
plt.figure(figsize=(12, 5))

# Plot Test Accuracy
plt.subplot(1, 2, 1)
plt.plot(d_values, test_accuracies, marker='o', color='blue')
plt.xlabel("Depth (d)")
plt.ylabel("Test Accuracy")
plt.title("Test Accuracy vs. Depth in Decision Tree")

# Plot Test F1 Score
plt.subplot(1, 2, 2)
plt.plot(d_values, test_f1_scores, marker='o', color='orange')
plt.xlabel("Depth (d)")
plt.ylabel("Test F1 Score")
plt.title("Test F1 Score vs. Depth in Decision Tree")

plt.tight_layout()
plt.show()
```
